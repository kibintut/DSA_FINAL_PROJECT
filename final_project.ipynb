# RECURSIVE AND ITERATION

```python
# Real World Problem: Let's say a user would like to create a large number of blank pages.
# It would be a hassle if the user will click the + button everytime to create a new page.

import time
import sys

# To Check Recursion Limit
print("Recursion Limit:")
print(sys.getrecursionlimit())
print("")

# Recursion
def create(numPages): 
    if numPages == 0:   # If numPage is less than or equal to 0, return numPages.
        return 
    create(numPages - 1)    # numPages - 1 until it reaches 0.
    print(f"Page {numPages} created!")  # Print Output

# Iteration
def Create(num_Pages):
    for created in range(1, num_Pages + 1): # for created in range from 1 to numPages + 1. 
        print(f"Page {created} created!")   # Print Output 


recTime = time.time()   # Record current time for recursion.
create(5)   # Execute function
print("Recursion Time: ", str(time.time()-recTime)) # Print out current time after execution - previous current recursion time. 

print("")

itTime = time.time() # Record current time for Iteration.
Create(5)   # Execute function
print("Iteration Time:", str(time.time()-itTime))   # Print out current time after execution - previous current Iteration time.
print("")

print("The advantages of Iteration over Recursion:" 
      "\nIteration is faster and more efficient than recursion." 
      ,"\nIteration is also more memory efficient than recursion.",
      "\nWhile Recursion has its limits , Iteration can handle Large inputs without any issues.")
```

```python
#Real World problem: Sorting the students based from their last name 

import time

#Data sets
small_data = ["Taylor", "Garcia", "Harris", "Perez", "Johnson", "Smith", 
              "Brown", "Lewis", "Zimmerman", "Robinson", "Clark", "Young", 
              "Nelson", "Martinez", "Anderson", "Underwood", "King", "Vargas", 
              "Foster", "Evans", "Olson", "Quinn", "Xander", "Williams"]

large_data =["Agripa", "Andueza","Perez", "Burito", "Jimenez", "Tablegi", "Kupal", "Popol",
              "Korpus", "Shalam", "Difeta", "Chupelagiti", "Samonivitch", "Milankovitch", "Xuxu",
              "Eguls", "Fuego", "Gigolo", "Habibi", "Ibraham", "Jaycool", "Luma","Nuger", "Opa", "Pipeto",
              "Qpal", "Ragaak", "Ugago", "Vrofetchto", "Wangwang","Xamel", "Yongo", "Zoro", "Abubi","Caranuni", "Jopay", "Yameti", "Shipudey", "Puwas", "Mimi", "Pepol","Frexa","Nolan",
               "Layla", "Zilong", "Buritto", "Miya", "Khulifa","Zonrox", "Qty", "Kalimot", "Snoppy", "Zed",
               "Duday", "Spike","Ewak", "Fita", "Gogol", "Corazon", "Verga", "Hanabishi", "Tyago", "Deldocruz",
               "Owawi", "Nyelepon", "Lapugo","Virgo", "Mendoza", "Bautista", "Wong", "Navarro", "Flores", "Reyes", "Aquino", 
                 "Uy", "Dela Cruz", "Garcia", "Santos", "Cruz", "Torres", "Ignacio", 
                 "Zaragoza", "Valdez", "Jimenez", "Katipunan", "Espino", "Lopez", 
                 "Perez", "Yabut", "Hernandez", "Quisumbing", "Xie"]

# Insertion Sort Function
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Quick Sort Function
def quick_sort(arr):        
    if len(arr) <= 1:        
        return arr          
    else:                   
        pivot = arr[0]      
        less = [x for x in arr[1:] if x <= pivot] 
        greater = [x for x in arr[1:] if x > pivot] 
        return quick_sort(less) + [pivot] + quick_sort(greater) 
    

# Time Complexity for Small Data Set using Insertion Sort
smallInsertionTime = time.time()
print(insertion_sort(small_data))
smallEndInsertionTime = str(time.time()-smallInsertionTime)
print("Sorting time for small data set using Insertion Sort: ", smallEndInsertionTime)
print("")

# Time Complexity for Small Data Set using Quick Sort
smallQuickTime = time.time()
print(quick_sort(small_data))
smallEndQuickTime = str(time.time()-smallQuickTime)
print("Sorting time for small data set using Quick Sort: ", smallEndQuickTime)
print("")


# Time Complexity for Large Data Set using Insertion Sort
largeInsertionTime = time.time()
print(insertion_sort(large_data))
largeEndInsertionTime = str(time.time()-largeInsertionTime)
print("Sorting time for large data set Insertion Sort: ", largeEndInsertionTime)
print("")

# Time Complexity for Large Data Set using Quick Sort
largeQuickTime = time.time()
print(quick_sort(large_data))
largeEndQuickTine = str(time.time()-largeQuickTime)
print("Sorting time for large data set using Quick Sort: ", largeEndQuickTine)
print("")

#Comparing Small Data Set Time Complexities for both Insertion Sort and Quick Sort
print("Comparing time complexities for small data set:")
if(smallEndInsertionTime < smallEndQuickTime):
    print("Insertion Sort is faster than Quick Sort")
elif(smallEndQuickTime < smallEndInsertionTime):
    print("Quick Sort is faster than Insertion Sort")
else:
    print("Insertion Sort and Quick Sort have the same complexity time for Small Data Set.")

print("")

#Comparing Large Data Set
print("Comparing time complexities for large data set:")
if(largeEndInsertionTime < largeEndQuickTine):
    print("Insertion Sort is faster than Quick Sort")
elif(largeEndQuickTine < largeEndInsertionTime):
    print("Quick Sort is faster than Insertion Sort")
else:
    print("Insertion Sort and Quick Sort have the same complexity time for Large Data Set.")
```

FIBONACCI PROBLEM

```python
# Fibonacci Problem

import time

n = 20

# # Recursive 
def fib(n):
    if n <= 1:  # Simply this is the base case if n is 0 or 1, we return n.
        return n    
    else:       # Else we call the function with n-1 and n-2 and add them together.
        return fib(n-1) + fib(n-2)

fibTime = time.time()   # Start time to record the time it takes to run the function.
print(f"The {n}th fibonacci number is: {fib(n)}") 
fibEndTime = str(time.time() - fibTime) # End time to record the time it takes to run the function.
print("Time Complexity for the Recursive approach: ", fibEndTime)    # Print the time it took to run the function.
print("")

# Top-Down Memoization
def fibMemo(n):
    memo = {0:0, 1:1}   # This is for creating a dictionary to store the results of the function calls.

    def f(x):
        if x in memo:   # Here we check if the value is already in the directory and if it is then we return it.
            return memo[x]
        else:
            memo[x] = f(x-1) + f(x-2)   # Else if it is not we calculate it and store it in the directory.
            return memo[x]
        
    return f(n)     # Return the result of the function.

fibMemoTime = time.time()  # Same as the previous time function to record the time it takes to run the function.
print(f"The {n}th fibonacci number is: {fibMemo(n)}")
fibMemoEndTime = str(time.time() - fibMemoTime)
print("Time Complexity for Top-Down Memoization: ", fibMemoEndTime)
print("")

# Bottom-Up Tabulation
def fibTab(n):

    if n == 0:          # Simply this is the base case if n is 0 then we return 0 and if n is 1 then we return 1. 
        return n
    if n == 1:
        return n
    
    dp = [0] * (n+1)        # This is for creating a list to store the result of the function calls.
    dp[0] = 0               
    dp[1] = 1

    for i in range(2, n+1):         # This is for looping through the list and calculating the result of the function calls.
        dp[i] = dp[i-2] + dp[i-1]   # Here we calculate the result of the function calls and store it in the list.
    
    return dp[n]                    # Here we return the result of the function calls.

fibTabTime = time.time()            # Same as the previous time function comments.
print(f"The {n}th fibonacci number is: {fibTab(n)}")
fibTabEndTime = str(time.time() - fibTabTime)
print("Time Complexity for Bottom-Up Tabulation: ", fibTabEndTime)
print("")

# Now we compare the time complexity of the dynamic programming solutions with the recursive solution. 
if(fibMemoEndTime, fibTabEndTime < fibEndTime):
    print("The Top-Down Memoization approach and the Bottom-Up Tabulation approach is faster than the Recursive approach.")
elif(fibMemoEndTime, fibTabEndTime == fibEndTime):
    print("The Recursive approach is as equal as the Top-Down Memoization approach and the Bottom-Up Tabulation approach.")
else:
    print("The Recursive approach is faster than the Top-Down Memoization and the Bottom-Up Tabulation approach.")

```

KNAPSACK PROBLEM

```python

# Knapsack Problem 

# Top-Down Memoization
def KSMemo(v, w, n, c, memo):

    if(n, c) in memo:
        return memo[(n, w)]
    if n == 0 or w == 0:
        return 0
    
    if w[n - 1] > c:
        memo[(n, c)] = KSMemo(v, w, n-1, c, memo)
    else:
        memo[(n, c)] = max(v[n - 1] + KSMemo(v, w, n - 1, c - w[n - 1], memo),
                           KSMemo(v, w, n - 1, c, memo))
    return memo[(n, c)]

def KS_with_memo(v, w, c):
    memo = {}
    return KSMemo(v, w, len(v), c, memo)

# Example Values
values = [60, 100, 120]
weights = [10, 20, 30]
c = 50

max_value_memo = KS_with_memo(values, weights, c)
print(max_value_memo)


# Bottom-Up Tabulation
def KSTab(v, w, c):
    n = len(v)
    dp = [[0 for _ in range(c + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(1, c + 1):
            if w[i - 1] <= j:
                dp[i][j] = max(dp[i - 1][j], v[i - 1] + dp[i - 1][j - w[i - 1]])
            else:
                dp[i][j] = dp[i - 1][j]

    return dp[n][c]

values = [60, 100, 120]
weights = [10, 20, 30]
c = 50

max_value_tabulation = KSTab(values, weights, c)
print(max_value_tabulation)
```

Graph Algorithm

```python
import heapq

def dijkstra(graph, start):
    min_distances = {node: float('infinity') for node in graph}
    min_distances[start] = 0
    pq = [(0, start)]

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        if current_distance > min_distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < min_distances[neighbor]:
                min_distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return min_distances

# Graph represented as adjacency list with weights
graph = {
    'Lansing, MI': {'Columbus, OH': 254},
    'Columbus, OH': {'Lansing, MI': 254, 'Charleston, WV': 162, 'Frankfort, KY': 186, 'Indianapolis, IN': 175},
    'Charleston, WV': {'Columbus, OH': 162, 'Frankfort, KY': 198},
    'Frankfort, KY': {'Charleston, WV': 198, 'Columbus, OH': 186, 'Indianapolis, IN': 166},
    'Indianapolis, IN': {'Columbus, OH': 175, 'Frankfort, KY': 166}
}

# Start the algorithm from 'Lansing, MI'
shortest_paths = dijkstra(graph, 'Frankfort, KY')
print(shortest_paths)
```
